```python
import pandas as pd
import numpy as np
```

```python
my_df = pd.read_csv('music.csv')
my_df
```

```python
genders = ['male', 'female', 'male', 'female', 'male', 'female']
my_df["Genders"] = genders  #Adding a new column in dataset
my_df
```

```python
my_df['Alive/Dead'] = [True]*len(my_df)  #Another way to add a column
my_df
```

```python
my_df['Show Dog'] = [np.nan]*len(my_df)  #Adding a null value
my_df
```

```python
my_df.insert(1, 'Good Person?', [True]*len(my_df))  #With insert function a column can be added at any position
my_df 
```
